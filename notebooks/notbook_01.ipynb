```python
def ma_strategy(df):
    df = df.copy()

    # 1. Signal: MA crossover
    df['signal'] = (df['ma_60'] >= df['ma_90']).astype(int)

    # 2. Position: shift signal (trade ở ngày sau)
    df['position'] = df['signal'].shift(1).fillna(0)

    # 3. Strategy return
    df['strategy_return'] = df['position'] * df['return_ratio']

    # 4. Equity curve
    df['equity'] = (1 + df['strategy_return']).cumprod()

    return df
df = ma_strategy(df)
```
